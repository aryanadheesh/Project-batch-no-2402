```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

homepage_url = "http://statisticstimes.com/"

homepage_response = requests.get(homepage_url)

homepage_soup = BeautifulSoup(homepage_response.text, "html.parser")

economy_link = homepage_soup.find("div", class_="dropdown-content").a["href"]

economy_page_url = homepage_url + economy_link

economy_page_response = requests.get(economy_page_url)

economy_page_soup = BeautifulSoup(economy_page_response.text, "html.parser")

state_gdp_link = economy_page_soup.find("a", string="» Economy » Indian States GSDP")

if state_gdp_link:
    state_gdp_link = state_gdp_link.get("href")

    state_gdp_page_url = homepage_url + state_gdp_link

    state_gdp_page_response = requests.get(state_gdp_page_url)

    state_gdp_page_soup = BeautifulSoup(state_gdp_page_response.text, "html.parser")

    gdp_table = state_gdp_page_soup.find("table", id="table_id")

    rank_list = []
    state_list = []
    gsdp_18_19_list = []
    gsdp_19_20_list = []
    share_18_19_list = []
    gdp_billion_list = []

    for row in gdp_table.find_all("tr")[1:]:
        columns = row.find_all("td")
        rank = columns[0].text.strip()
        state = columns[1].text.strip()
        gsdp_18_19 = columns[2].text.strip()
        gsdp_19_20 = columns[3].text.strip()
        share_18_19 = columns[4].text.strip()
        gdp_billion = columns[5].text.strip()

        rank_list.append(rank)
        state_list.append(state)
        gsdp_18_19_list.append(gsdp_18_19)
        gsdp_19_20_list.append(gsdp_19_20)
        share_18_19_list.append(share_18_19)
        gdp_billion_list.append(gdp_billion)

    df = pd.DataFrame({
        "Rank": rank_list,
        "State": state_list,
        "GSDP(18-19) at Current Prices": gsdp_18_19_list,
        "GSDP(19-20) at Current Prices": gsdp_19_20_list,
        "Share(18-19)": share_18_19_list,
        "GDP($ billion)": gdp_billion_list
    })


    print(df)
else:
    print("State-wise GDP link not found on the Economy page.")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table", class_="wikitable")

if table:
   
    ranks = []
    names = []
    artists = []
    upload_dates = []
    views = []

    for row in table.find_all("tr")[1:]:
        columns = row.find_all("td")
   
        if len(columns) >= 5:
         
            rank = columns[0].text.strip()
            ranks.append(rank)

            name = columns[1].text.strip()
            names.append(name)

            artist = columns[2].text.strip()
            artists.append(artist)

            upload_date = columns[4].text.strip()
            upload_dates.append(upload_date)

            view = columns[3].text.strip()
            views.append(view)

    for i in range(len(ranks)):
        print(f"A) Rank: {ranks[i]}")
        print(f"B) Name: {names[i]}")
        print(f"C) Artist: {artists[i]}")
        print(f"D) Upload Date: {upload_dates[i]}")
        print(f"E) Views: {views[i]}")
        print("=" * 50)
else:
    print("Table containing the most viewed YouTube videos not found.")
```

```python
import requests
from bs4 import BeautifulSoup

homepage_url = "https://github.com/"

response = requests.get(homepage_url)

soup = BeautifulSoup(response.text, "html.parser")

explore_menu = soup.find("nav", class_="mt-0 flex-shrink-0")

if explore_menu:
    trending_link = explore_menu.find("a", string="Trending")
   
    if trending_link:
        trending_link = trending_link["href"]
       
        trending_page_url = homepage_url + trending_link

        response = requests.get(trending_page_url)

        soup = BeautifulSoup(response.text, "html.parser")

        repositories = soup.find_all("article", class_="Box-row")

        repo_titles = []
        descriptions = []
        contributors_counts = []
        languages = []

        for repo in repositories:

            repo_title = repo.h1.a.text.strip()
            repo_titles.append(repo_title)

            description = repo.find("p", class_="col-9 color-text-secondary my-1 pr-4")
            description_text = description.text.strip() if description else ""
            descriptions.append(description_text)

            contributors = repo.find("a", class_="Link--muted d-inline-block mr-3")
            contributors_count = contributors.text.strip() if contributors else "0"
            contributors_counts.append(contributors_count)

            language = repo.find("span", itemprop="programmingLanguage")
            language_text = language.text.strip() if language else ""
            languages.append(language_text)

        for i in range(len(repo_titles)):
            print("Repository Title:", repo_titles[i])
            print("Repository Description:", descriptions[i])
            print("Contributors Count:", contributors_counts[i])
            print("Language Used:", languages[i])
            print("-----------------------------------------")
    else:
        print("Trending link not found.")
else:
    print("Explore menu not found.")
```

```python
import requests
from bs4 import BeautifulSoup

homepage_url = "https://www.billboard.com/"

response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

charts_link = soup.find("a", class_="header__main-link header__main-link--charts")

if charts_link:
 
    charts_page_url = charts_link["href"]

    response = requests.get(charts_page_url)

    soup = BeautifulSoup(response.text, "html.parser")

    hot_100_link = soup.find("a", class_="charts-landing__link charts-landing__video-link")

    if hot_100_link:
       
        hot_100_page_url = hot_100_link["href"]

      
        response = requests.get(hot_100_page_url)

        soup = BeautifulSoup(response.text, "html.parser")

        songs = soup.find_all("li", class_="chart-list__element")

        song_names = []
        artist_names = []
        last_week_ranks = []
        peak_ranks = []
        weeks_on_board = []

        for song in songs:
      
            song_name = song.find("span", class_="chart-element__information__song text--truncate color--primary").text.strip()
            song_names.append(song_name)

            artist_name = song.find("span", class_="chart-element__information__artist text--truncate color--secondary").text.strip()
            artist_names.append(artist_name)

            last_week_rank = song.find("span", class_="chart-element__meta text--center color--secondary text--last").text.strip()
            last_week_ranks.append(last_week_rank)

            peak_rank = song.find("span", class_="chart-element__meta text--center color--secondary text--peak").text.strip()
            peak_ranks.append(peak_rank)
            weeks_on_board = song.find("span", class_="chart-element__meta text--center color--secondary text--week").text.strip()
            weeks_on_board.append(weeks_on_board)

        for i in range(len(song_names)):
            print("Song Name:", song_names[i])
            print("Artist Name:", artist_names[i])
            print("Last Week Rank:", last_week_ranks[i])
            print("Peak Rank:", peak_ranks[i])
            print("Weeks on Board:", weeks_on_board[i])
            print("-----------------------------------------")
    else:
        print("Hot 100 page link not found.")
else:
    print("Charts section link not found.")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table")

book_names = []
author_names = []
volumes_sold = []
publishers = []
genres = []

for row in table.find_all("tr")[1:]:  
    columns = row.find_all("td")
    
    if len(columns) >= 5: 
        
        book_name = columns[0].text.strip()
        book_names.append(book_name)

        author_name = columns[1].text.strip()
        author_names.append(author_name)

        volumes = columns[2].text.strip()
        volumes_sold.append(volumes)

        publisher = columns[3].text.strip()
        publishers.append(publisher)

        genre = columns[4].text.strip()
        genres.append(genre)

for i in range(len(book_names)):
    print(f"Book Name: {book_names[i]}")
    print(f"Author Name: {author_names[i]}")
    print(f"Volumes Sold: {volumes_sold[i]}")
    print(f"Publisher: {publishers[i]}")
    print(f"Genre: {genres[i]}")
    print("=" * 50)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.imdb.com/list/ls512407256/"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

tv_series_section = soup.find("div", class_="lister-list")

if tv_series_section:
   
    names = []
    year_spans = []
    genres = []
    runtimes = []
    ratings = []
    votes = []

    for series in tv_series_section.find_all("div", class_="lister-item-content"):
    
        name_element = series.find("h3", class_="lister-item-header")
        if name_element:
            name = name_element.find("a").text.strip()
        else:
            name = "N/A"
        names.append(name)

        year_span_element = series.find("span", class_="lister-item-year")
        if year_span_element:
            year_span = year_span_element.text.strip()
        else:
            year_span = "N/A"
        year_spans.append(year_span)

        genre_element = series.find("span", class_="genre")
        if genre_element:
            genre = genre_element.text.strip()
        else:
            genre = "N/A"
        genres.append(genre)

        runtime_element = series.find("span", class_="runtime")
        if runtime_element:
            runtime = runtime_element.text.strip()
        else:
            runtime = "N/A"
        runtimes.append(runtime)

        rating_element = series.find("div", class_="ipl-rating-star")
        if rating_element:
            rating = rating_element.find("span", class_="ipl-rating-star__rating").text.strip()
        else:
            rating = "N/A"
        ratings.append(rating)

        vote_element = series.find("p", class_="sort-num_votes-visible")
        if vote_element:
            vote = vote_element.find("span", attrs={"name": "nv"}).text.strip()
        else:
            vote = "N/A"
        votes.append(vote)

    for i in range(len(names)):
        print(f"Name: {names[i]}")
        print(f"Year Span: {year_spans[i]}")
        print(f"Genre: {genres[i]}")
        print(f"Run Time: {runtimes[i]}")
        print(f"Rating: {ratings[i]}")
        print(f"Votes: {votes[i]}")
        print("=" * 50)
else:
    print("TV series section not found.")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.bcci.tv/"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

fixture_link = None
for link in soup.find_all("a", href=True):
    if link.text.strip() == "INTERNATIONAL FIXTURES":
        fixture_link = link["href"]
        break

if fixture_link:
    fixture_url = url + fixture_link

    fixture_response = requests.get(fixture_url)
    fixture_soup = BeautifulSoup(fixture_response.text, "html.parser")

    fixtures_section = fixture_soup.find("ul", class_="js-list")

    if fixtures_section:
      
        series = []
        places = []
        dates = []
        times = []

        for fixture in fixtures_section.find_all("li"):
  
            series_text = fixture.find("strong").text.strip()
            series.append(series_text)

            place = fixture.find("p", class_="fixture__additional-info").text.strip()
            places.append(place)

            date = fixture.find("span", class_="fixture__date").text.strip()
            dates.append(date)

            time = fixture.find("span", class_="fixture__time").text.strip()
            times.append(time)

        for i in range(len(series)):
            print(f"A) Series: {series[i]}")
            print(f"B) Place: {places[i]}")
            print(f"C) Date: {dates[i]}")
            print(f"D) Time: {times[i]}")
            print("=" * 50)
    else:
        print("Fixtures section not found.")
else:
    print("Link to international fixtures page not found.")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://archive.ics.uci.edu/"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

link = soup.find("a", href="datasets.php")

if link:

    all_datasets_url = url + link["href"]

    all_datasets_response = requests.get(all_datasets_url)
    all_datasets_soup = BeautifulSoup(all_datasets_response.text, "html.parser")

    dataset_table = all_datasets_soup.find("table", class_="table")

    if dataset_table:
    
        dataset_names = []
        data_types = []
        tasks = []
        attribute_types = []
        no_of_instances = []
        no_of_attributes = []
        years = []

        for row in dataset_table.find_all("tr")[1:]:
            columns = row.find_all("td")

            if len(columns) >= 7:
              
                dataset_name = columns[0].text.strip()
                dataset_names.append(dataset_name)

                data_type = columns[1].text.strip()
                data_types.append(data_type)

                task = columns[2].text.strip()
                tasks.append(task)

                attribute_type = columns[3].text.strip()
                attribute_types.append(attribute_type)

                no_instances = columns[4].text.strip()
                no_of_instances.append(no_instances)

                no_attributes = columns[5].text.strip()
                no_of_attributes.append(no_attributes)

                year = columns[6].text.strip()
                years.append(year)

        for i in range(len(dataset_names)):
            print(f"A) Dataset Name: {dataset_names[i]}")
            print(f"B) Data Type: {data_types[i]}")
            print(f"C) Task: {tasks[i]}")
            print(f"D) Attribute Type: {attribute_types[i]}")
            print(f"E) No of Instances: {no_of_instances[i]}")
            print(f"F) No of Attributes: {no_of_attributes[i]}")
            print(f"G) Year: {years[i]}")
            print("=" * 50)
    else:
        print("Dataset table not found.")
else:
    print("Link to Show All Dataset page not found.")
```
